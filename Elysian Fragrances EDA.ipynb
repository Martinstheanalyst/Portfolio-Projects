```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv("df_elysian_cleaned.csv",
              index_col='Sales ID')
df
```

Adding a Sales Column

```python
df['Sales']=df['Price']*df['Quantity']
```

```python
df
```

Profit Column

```python
df['Profit']=(df['Price']-df['Cost Price'])-df['Discount']*df['Quantity']
```

```python
df
```

Lets find the highest selling products and also the most profitable products

```python
#lets disable the scientific notation
pd.options.display.float_format = '{:.2f}'.format
```

```python
sales_by_product=df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
sales_by_product
#from the output, we can see that aqua has the highest sales over the past 5years followed by Belle, Dusk etc respectively.
```

```python
profit_by_product=df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False)
profit_by_product
```

Lets find the total number of quantities sold during the past 5 years


```python
total_qty_sold=df['Quantity'].sum()
total_qty_sold
```

Highest selling Bottle Sizes

```python
no_bottle_size_sold=df.groupby('Bottle Size (ml)')['Sales'].count().sort_values(ascending=False)
no_bottle_size_sold
```

Top  selling Categories with profits

```python
sum_category=df.groupby('Category')[['Sales','Profit']].sum().sort_values(by='Profit', ascending=False)
sum_category
```

Cities

```python
Salesprofit_by_cities=df.groupby('City')[['Sales','Profit']].sum().sort_values(by='Profit', ascending=False)
Salesprofit_by_cities
```

User Type and Gender

```python
total_sales=round(sum(df['Sales']),2)
total_sales
```

```python
total_profit=round(sum(df['Profit']),2)
total_profit
```

```python
sales_by_user_type=df.groupby('User Type')['Sales'].sum().sort_values(ascending=False)
sales_by_user_type
```

```python
perc_sales_by_usertype=sales_by_user_type/total_sales *100
perc_sales_by_usertype
```

```python
sales_by_gender=df.groupby('Gender')[['Sales', 'Profit']].sum()
sales_by_gender
```

```python
sales_by_gender/total_sales *100
```

```python
#from the two outputs above, we could see that the usertypes designated fr  each product has those for women as the highest sold
#and unisex following next. Howver, when it comes to the gender doing the purchasses, we could see that Males have the...
#highest sales and profit. meaning that more men are more interested in female scents and unisex scents that male scents.
#this repreents only a tiny percentage.
```

Correlation

Let's find the correlation for the sales, bottle size, city columns in the dataset

```python
correlation_matrix=df[['Sales', 'Bottle Size (ml)']].corr()
correlation_matrix

#the output shows that there is no correlation between sales and bottle size. i.e there is no linear relationship between them.
```

```python
#discount correlation
df[['Sales', 'Discount']].corr()

#the output shows a weak positive relationship between sales and discount. because the correlation is still close to 0.
#this means that changes in discount have no impact  on sales 
```

```python
df['Price'].corr(df['Quantity'])

#output shows a weak negative corelation between price and quantity. there is no linear relationship between them. this
#could mean there is a tendancy for higher prices to be associated with slightly lower prices.
```
